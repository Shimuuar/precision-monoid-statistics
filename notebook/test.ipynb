```python
%matplotlib widget
import okaflow
import numpy as np
import matplotlib.pyplot as plt
import h5py as h5
```

```python
arg = okaflow.Args.fromEnv()
arg
```

```python
meta = okaflow.fromEnv()
meta
```

```python
paths = okaflow.parse(okaflow.Many(okaflow.Path('data.h5')), arg)
paths
```

```python
def loadH5(p:str):
    with h5.File(p) as f:
        return f['dat'][:]
```

```python
dat = [loadH5(p) for p in paths]
```

```python
plt.figure()
for d in dat:
    plt.hist(d['est']-d['exact'], density=True, bins=50, histtype='step')
None
```
